```python
import shutil

cnt=0
for filename in files:
    f_name = filename.split('.')
    
    if f_name[1] == "txt":
        new_f = f_name[0] + '_flip.'+f_name[1]
        shutil.copy(output_dir+filename, output_dir+new_f)
        print(new_f)

        f_r = open(filename,'r+')
        f_text = f_r.read().split(' ')
        print(f_text)
                
        f_r.close()
        
        x = 1 - float(f_text[1])
        
        f_w = open(new_f,'w+')
        fw_text = f_text[0] + ' ' + str(x) + ' ' + f_text[2] + ' ' + f_text[3] + ' ' + f_text[4]
#        print(fw_text)
        f_w.write(fw_text)
                
        f_w.close()


#        print('txt : ' + filename)
#        cnt += 1
#        if cnt == 3 :
#            break
        
    
    else :
        original = cv2.imread(filename,cv2.IMREAD_COLOR)
        flip_img=cv2.flip(original,1)
        cv2.imwrite(f_name[0] + '_flip.' + f_name[1], flip_img)
#        cv2.imshow("oringinal",original)
#        cv2.imshow("flip",flip_img)
#        print(filename)
       
```
